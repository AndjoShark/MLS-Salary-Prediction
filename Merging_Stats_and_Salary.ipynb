```python
import pandas as pd
import numpy as np
stats = pd.read_json('MLS_stats_final.json')
salary = pd.read_csv('salaries.csv')
```

```python
stats = stats[stats['Club'] != '-']
stats = stats.sort_values(by = ['url', 'Year'])
stats = stats.dropna()
stats.shape
```

```python
salary['url'] = salary['urls']
urls = salary['url']
salary.head().T
```

```python
non_aggregate_cols = ['Year', 'Club', 'url']
aggregate_cols = [cols for cols in stats.columns if cols not in non_aggregate_cols]
import numpy as np
stats_aggregate = stats.groupby(non_aggregate_cols).agg(np.sum).reset_index()
```

```python
stats_aggregate.shape
```

```python
stats.shape
```

```python
stats_aggregate['previous_year'] = stats_aggregate['Year'].astype(int)
```

```python
salary['previous_year'] = salary['season'] - 1
df = pd.merge(stats_aggregate, salary, on= ['previous_year', 'url'])
```
