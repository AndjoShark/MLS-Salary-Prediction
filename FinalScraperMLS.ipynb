```python
import scrapy
from scrapy.crawler import CrawlerProcess
class MLS_Spider(scrapy.Spider):
    
    name = "EvenStats"
    download_delay = 1.0
    start_urls = urls

    def parse(self, response):
        
        for odd in response.xpath('//tr[@class="odd"]'):
            yield {
                'url': response.request.url,
                'Year': odd.xpath('td[@data-title="Year"]/text()').extract_first(),
                'Club': odd.xpath('td[@data-title="Club"]/text()').extract_first(),
                'GP': odd.xpath('td[@data-title="GP"]/text()').extract_first(),
                'GS': odd.xpath('td[@data-title="GS"]/text()').extract_first(),
                'G': odd.xpath('td[@data-title="G"]/text()').extract_first(),
                'MINS': odd.xpath('td[@data-title="MINS"]/text()').extract_first(),
                'A': odd.xpath('td[@data-title="A"]/text()').extract_first(),
                'SHTS': odd.xpath('td[@data-title="SHTS"]/text()').extract_first(),
                'SOG': odd.xpath('td[@data-title="SOG"]/text()').extract_first(),
                'FC': odd.xpath('td[@data-title="FC"]/text()').extract_first(),
                'OFF': odd.xpath('td[@data-title="OFF"]/text()').extract_first(),
                'Y': odd.xpath('td[@data-title="Y"]/text()').extract_first(),
                'R': odd.xpath('td[@data-title="R"]/text()').extract_first()
            
              }
            
        for odd in response.xpath('//tr[@class="even"]'):
            yield {
                'url': response.request.url,
                'Year': odd.xpath('td[@data-title="Year"]/text()').extract_first(),
                'Club': odd.xpath('td[@data-title="Club"]/text()').extract_first(),
                'GP': odd.xpath('td[@data-title="GP"]/text()').extract_first(),
                'GS': odd.xpath('td[@data-title="GS"]/text()').extract_first(),
                'G': odd.xpath('td[@data-title="G"]/text()').extract_first(),
                'MINS': odd.xpath('td[@data-title="MINS"]/text()').extract_first(),
                'A': odd.xpath('td[@data-title="A"]/text()').extract_first(),
                'SHTS': odd.xpath('td[@data-title="SHTS"]/text()').extract_first(),
                'SOG': odd.xpath('td[@data-title="SOG"]/text()').extract_first(),
                'FC': odd.xpath('td[@data-title="FC"]/text()').extract_first(),
                'OFF': odd.xpath('td[@data-title="OFF"]/text()').extract_first(),
                'Y': odd.xpath('td[@data-title="Y"]/text()').extract_first(),
                'R': odd.xpath('td[@data-title="R"]/text()').extract_first()
            
              }


        
        
            

process = CrawlerProcess({
    'FEED_FORMAT': 'json',         # Store data in JSON format.
    'FEED_URI': 'MLS_stats_final.json',  # Name our storage file.
    'LOG_ENABLED': False           # Turn off logging for now.
})

# Start the crawler with our spider.
process.crawl(MLS_Spider)
process.start()
print('Success!')
```
